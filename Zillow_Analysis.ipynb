```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
# @title

from enum import Enum
import pandas as pd
import pandas_bokeh
import csv
import requests
from bokeh.models.tickers import SingleIntervalTicker
from os.path import exists
from os import mkdir
from google.colab import files


class WebFetcherWithCache:

    def __init__(self):
        self.cacheDirectory = './cached-data/'

    def get(self, url):
        fileName = url.split('/')[-1]
        cachePath = self.cacheDirectory + fileName
        if exists(cachePath):
            return pd.read_csv(cachePath)
        else:
            data = None
            try:
                data = pd.read_csv(url)
            except:
                return None
            data.to_csv(cachePath)
            return data


class HousingMarketReport:

    def __init__(
        self,
        location=None,
        reportName='ZillowAnalysis.html',
        reportStartDate='2018-01-01',
        skipSmoothedData=True,
        ):
        self.reportStartDate = reportStartDate
        self.reportMaxDate = '2024-01-01'
        self.reportName = reportName
        self.plots = []
        self.fetcher = WebFetcherWithCache()
        self.skipSmoothedData = skipSmoothedData
        self.location = location

    def __generatePlotsForCategory(self, categoryName,
                                   overrideLocation):
        location = overrideLocation or self.location
        if not location:
            raise ValueError('Missing location')

        categoryData = data[categoryName]
        for (title, subCategories) in categoryData.items():
            if self.skipSmoothedData and 'Raw' not in title:  # Only look at raw data, not smoothed
                continue
            dataUrl = subCategories.get('Metro & U.S. ') \
                or subCategories.get('Metro & U.S.')
            if not dataUrl:
                continue

            df1 = self.fetcher.get(dataUrl)
            if df1 is None:
                continue

            df1 = df1[df1['RegionName'].str.contains(location)]
            region = df1['RegionName'].to_string(index=False)

            df1.drop([
                'RegionID',
                'SizeRank',
                'RegionType',
                'RegionName',
                'StateName',
                'BaseDate',
                ], inplace=True, axis=1, errors='ignore')
            df1 = df1.transpose().reset_index()
            if len(df1.columns) != 2:
                continue

            df1.rename(columns={df1.columns[0]: 'Date',
                       df1.columns[1]: 'Value'}, inplace=True)

            df1 = df1[df1['Date'] >= self.reportStartDate]
            df1 = df1[df1['Date'] < self.reportMaxDate]

            plotTitle = title + ' in ' + region
            figure = df1.plot_bokeh.line(
                title=plotTitle,
                show_figure=False,
                x='Date',
                y='Value',
                figsize=(900, 300),
                legend=False,
                )
            figure.yaxis.formatter.use_scientific = False
            figure.xaxis.ticker = SingleIntervalTicker(interval=10,
                    desired_num_ticks=10)

            self.plots.append(figure)
            return self

    def build(self):
        pandas_bokeh.output_file(self.reportName)
        pandas_bokeh.plot_grid(self.plots, sizing_mode='scale_width',
                               ncols=2, show_plot=True)
        files.download(self.reportName)

    def addHomeValuesSection(self, overrideLocation=None):
        return self.__generatePlotsForCategory('HOME VALUES',
                overrideLocation)

    def addHomeValuesForecastSection(self, overrideLocation=None):
        return self.__generatePlotsForCategory('HOME VALUES FORECASTS',
                overrideLocation)

    def addInventorySection(self, overrideLocation=None):
        return self.__generatePlotsForCategory('INVENTORY',
                overrideLocation)

    def addListAndSalePricesSection(self, overrideLocation=None):
        return self.__generatePlotsForCategory('LIST AND SALE PRICES',
                overrideLocation)

    def addSalesCountAndPriceCutsSection(self, overrideLocation=None):
        return self.__generatePlotsForCategory('SALES COUNT AND PRICE CUTS'
                , overrideLocation)


def start():
    report = HousingMarketReport(location='San Diego')
    report.addHomeValuesSection().addInventorySection().addListAndSalePricesSection().build()


start()
```
